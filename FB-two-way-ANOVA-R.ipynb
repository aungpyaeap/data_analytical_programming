```r
gender <- c('Male','Female','Male','Female','Male','Female','Male','Female','Male','Female')
post <- c('P1','P1','P2','P2','P3','P3','P4','P4','P5','P5')
result <- c(4,14,2,6,0,1,2,5,20,40)
#startdate <- as.Date(c('2010-11-1','2008-3-25','2007-3-14'))

facebook <- data.frame(gender, post, result)
facebook
```

```r
library(ggplot2)
ggplot(facebook, aes(x=as.factor(post), y=result)) + geom_boxplot(fill="orange", alpha=0.2) + 
  xlab("Post") + ylab("Result") + ggtitle("Boxplot grouped by Post")
```

```r
ggplot(facebook, aes(x=gender, y=result)) + 
  geom_boxplot(color="red", fill="orange", alpha=0.2) +
  xlab("Gender") + ylab("Result") + ggtitle("Boxplot grouped by Gender")
```

```r
summary(facebook)
```

```r
anova_result <- aov(result ~ gender + post, data=facebook)
summary(anova_result)
```
